```python
titles = ['Wolf of Wall Street, The (2013)', 'Jumanji (1995)', 'Shawshank Redemption, The (1994)', 'Godfather, The (1972)']
filtered_df = merged_new_df[merged_new_df['title'].isin(titles)]

max_avg_rating_row = filtered_df.loc[filtered_df['rating_average'].idxmax()]

print("Title with the maximum average rating:")
print(max_avg_rating_row['title'])
```
